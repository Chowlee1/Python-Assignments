```python
mydb = pymysql.connect(
    host = 'localhost',
    user = 'root',
    port = int(3307),
    password = 'Utuama30',
    database = 'PatientManagement_system'
)

mycursor = mydb.cursor()
```

```python
class Patients_records:
    
    def displayRecords(self):
        mycursor.execute('select * from Patient_information')
        for x in mycursor:
            print(x)
            
    def acceptRecords(self):
        #accept records
        position = input('input position')
        namee = input('input name')
        sexx = input('input sex')
        bloodgroupp = input('blood group')
        genotypee = input('genotype')
        agee = input('age')

        sql = 'insert into Patient_information(patientID,patient_name,sex,blood_group,genotype,age) values(%s,%s,%s,%s,%s,%s)'
        val = [(int(position),namee,sexx,bloodgroupp,genotypee,int(agee)
               )]
        mycursor.executemany(sql,val)
        mydb.commit()
        
    def searchRecords(self):
        #search patient records
        search = input('what are you looking for?')
        mycursor.execute('select * from Patient_information where patient_name like "{}%" or sex like "{}%" or blood_group like "{}%" or genotype like "{}%" or age like "{}%" '.format(search,search,search,search,search))
        for x in mycursor:
            print(x)
            
    def updateRecords(self):
        # update patient records
        ID = input('what is the patient ID?')
        column = input('what column are you updating?')
        value = input('what value are you updating it to?')


        mycursor.execute('update Patient_information set {} = "{}" where patientID = "{}"'.format(column,value,ID))
        mycursor.execute('select * from Patient_information')
        for x in mycursor:
            print(x)
            
    def deleteRecords(self):
            
        #delete patient records
        Patient = input('what patient ID or name do you want to delete?')
        mycursor.execute('delete from Patient_information where patientID like "{}%" or patient_name like "{}%"'.format(Patient,Patient))
        mycursor.execute('select * from Patient_information')
        for x in mycursor:
            print(x)
            
    
    
   
    
            
        
```

```python
def start():
    
    program = Patients_records()
    
    print('''Hello! This program allows you to Display, Add, Search, Update and Delete records
    from the patient management database''')
    x = input('Enter "y" to begin and "n" to end your operation:')
    while x =='y':
        a = input('Would you like to display records? "y" for yes, "n" for no')
        while a =='y':
            program.displayRecords()
            a = input('Would you like to display another? "y" for yes, "n" for no')
        b = input('Would you like to input records? "y" for yes, "n" for no')
        while b =='y':
            program.acceptRecords()
            b = input('Would you like to add another? "y" for yes, "n" for no')
        c = input('Would you like to search records? "y" for yes, "n" for no')
        while c =='y':
            program.searchRecords()
            c = input('Would you like to search another? "y" for yes, "n" for no')
        d = input('Would you like to update records? "y" for yes, "n" for no')
        while d =='y':
            program.updateRecords()
            d = input('Would you like to update another? "y" for yes, "n" for no')
        e = input('Would you like to delete records? "y" for yes, "n" for no')
        while e =='y':
            program.deleteRecords()
            e = input('Would you like to delete another? "y" for yes, "n" for no')
        x = input('Enter "y" to continue and "n" to end your operation:')
```

```python
start()
```
