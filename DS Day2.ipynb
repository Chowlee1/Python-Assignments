```python
months = ['Jan', 'Feb', 'Mar','Apr','May','Jun','Jul','Aug','Sept','Oct','Nov','Dec']
month_index = np.arange(1,31)
np.random.seed(10)
vals = np.random.randint(5000,500001,360)
vals = vals.reshape(30,12)

calender = pd.DataFrame(data = vals, index= month_index, columns = months)
calender
```

```python
mean_sales = (f'The average sales for Jan is {int(np.mean(calender["Jan"]))}, for May is {int(np.mean(calender["May"]))}, for Aug is {int(np.mean(calender["Aug"]))} and for Sept is {int(np.mean(calender["Sept"]))}')

print(mean_sales)
```

```python
calender.loc['max_sales'] = np.max(calender.iloc[0:30,0:12]) #retrieving a subset of rows and columns
calender
```

```python
feb_percentile = (f'The 20th percentile for Feb is {np.percentile(calender["Feb"],20)}, the 50th is {np.percentile(calender["Feb"],50)}, and the 75th is {np.percentile(calender["Feb"],75)}.')
jun_percentile = (f'The 20th percentile for Jun is {np.percentile(calender["Jun"],20)}, the 50th is {np.percentile(calender["Jun"],50)}, and the 75th is {np.percentile(calender["Jun"],75)}.')
dec_percentile = (f'The 20th percentile for Dec is {np.percentile(calender["Dec"],20)}, the 50th is {np.percentile(calender["Dec"],50)}, and the 75th is {np.percentile(calender["Dec"],75)}.')

print(feb_percentile)
print('')
print(jun_percentile)
print('')
print(dec_percentile)
```

```python
total_salesJun = calender['Jun'].sum()
total_salesJul = calender['Jul'].sum()

print(f'The total sales in Jun was {total_salesJun}')
print(f'The total sales in Jul was {total_salesJul}')


if total_salesJun < total_salesJul:
    print(f'Difference between total sales in Jun and Jul is {total_salesJun - total_salesJul}, there were less sales in Jun compared to Jul')
elif total_salesJun > total_salesJul:
    print(f'Difference between total sales in Jun and Jul is {total_salesJun - total_salesJul}, there were more sales in Jun compared to Jul')
else:
    print(f'Difference between total sales in Jun and Jul is {total_salesJun - total_salesJul}, there were no differences in sales between Jun and Jul')
    
    
```
